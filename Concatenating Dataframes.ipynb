```python
import pandas as pd
import numpy as np
```

# Define Datasets

```python
df_a = pd.DataFrame({"Client_id":[10,12,34,58,59,74,14,59], "Name":["Peter", "Joseph", "Arthur", "Hector", \
                                                                   "Lea", "Alicia", "Meredith", "Lisa"]})
df_b = pd.DataFrame({"Client_id":[11,105,54,45,1010], "Name":["Philippe", "Ariana", "Perceval", "Judith", \
                                                                   "Peneloppe"]})
```

```python
df_a.head()
```

```python
df_b.head()
```

# Concatenate Datasets (rows axis)

```python
df_concat = pd.concat([df_a,df_b]) # by default, this is pd.concat([df_a,df_b], axis=0) (or axis="rows")
df_concat
```

# Concatenate Datasets (Columns axis)

Let's change use diffent columns names : 

```python
df_c = df_b.rename(columns={"Client_id":"Age","Name":"Best Friend"})
df_c.head()
```

```python
df_concat_cols = pd.concat([df_a,df_c], axis = 1)
df_concat_cols
```

If I hadn't changed the name of the columns, this would have happened : 

```python
df_concat_cols = pd.concat([df_a,df_b], axis = 1)
df_concat_cols
```

So the columns keep the same name without any error message. However, it is not very convenient.
